```python

import os
import pandas
import codecs
import glob
import pandas as pd
 
 
os.getcwd()
os.chdir(r"C:\Users\18895\Desktop\164")
 
def txtcombine():
    
    files = glob.glob('*.txt')
    
    all = codecs.open('all.txt','a')
    
    for filename in files:  
        print(filename)  
        fopen=codecs.open(filename,'r',encoding='utf-8')  
        lines=[]  
        lines=fopen.readlines()  
        fopen.close()  
        i=0  
        for line in lines:    
            for x in line:  
                all.write(x)
        #读取为DataFrame格式
        all1 = pd.read_csv('all.txt',sep=' ',encoding='GB2312')
        #保存为csv格式
        all1.to_csv('all1.csv',encoding='GB2312')
        
if __name__ == '__main__':  
    txtcombine()
```
